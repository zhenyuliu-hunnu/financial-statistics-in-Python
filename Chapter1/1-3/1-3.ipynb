# 贵州茅台股价

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['Noto Sans SC']  # 用来正常显示中文标签，如果报错请安装字体「思源黑体」
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

# 读取数据
data = pd.read_csv("GZMTp.csv")
data['DATE'] = pd.to_datetime(data.iloc[:, 0])
dat = pd.Series(data.iloc[:, 1].values, index=data['DATE'])
```

```python
# 计算对数收益率
logR = np.log(dat)
logr = logR.diff()
logreturn = logr.dropna()

# 创建图表
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# 绘制日收盘价图
ax1.plot(dat.index, dat.values, linewidth=1)
ax1.set_xlabel("贵州茅台日收盘价")
ax1.set_ylabel("")
ax1.set_title("")

# 绘制对数收益率图
ax2.plot(logreturn.index, logreturn.values, linewidth=1)
ax2.set_xlabel("贵州茅台对数收益率")
ax2.set_ylabel("")
ax2.set_title("")

plt.tight_layout()
plt.show()
```
